```python
df = df[['Restaurant Name','Average Cost for two','Aggregate rating','Rating color','Rating text','Votes']]
```

```python
df.drop(['Rating color','Rating text','Votes'],inplace=True,axis=1)
```

```python
restaurants = {}
reverse_restaurants = {}
for i in range(0,len(df)):
    name = df.iloc[i]['Restaurant Name']
    restaurants[name] = i+1
    reverse_restaurants[i+1] = name
```

```python
restaurants
```

```python
reverse_restaurants
```

```python
similarity_matrix1 = np.zeros(shape=(20,20))
```

```python
X = np.array(df.drop(['Restaurant Name'],axis=1))
```

```python
X
```

```python
for i in range(0,X.shape[0]):
    for j in range(0,X.shape[0]):
        if i==j:
            similarity_matrix1[i,j] = 0
        else:
            similarity_matrix1[i,j] = np.sqrt((X[i,0]-X[j,0])*(X[i,0]-X[j,0]) + (X[i,1]-X[j,1])*(X[i,1]-X[j,1]))
```

```python
similarity_matrix1
```

```python
similarity_matrix1 = np.max(similarity_matrix1) - similarity_matrix1
```

```python
similarity_matrix1
```

```python
restaurants_ratings = {}
for name in restaurants.keys():
    restaurants_ratings[name] = df.iloc[restaurants[name]-1]['Aggregate rating']
```
