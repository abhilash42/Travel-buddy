```python
similarity_matrix = np.zeros(shape=(971,971))
```

```python
similarity_matrix
```

```python
X = np.array(final.drop(['hotels_index','city'],axis=1))
```

```python
X.shape
```

```python
for i in range(0,X.shape[0]):
    for j in range(0,X.shape[0]):
        if i == j:
            similarity_matrix[i][j] = 0.0
        else:
            sm = 0.0
            for k in range(0,X.shape[1]):
                sm = sm + (X[i,k]-X[j,k])*(X[i,k]-X[j,k])
            similarity_matrix[i][j] = np.sqrt(sm)
```
